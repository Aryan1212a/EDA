## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Data Cleaning and Preprocessing
print("\n=== DATA WRANGLING ===")

# Create a copy for cleaning
df_clean = df.copy()

# Handle missing values
# Fill missing directors with 'Unknown'
df_clean['director'] = df_clean['director'].fillna('Unknown')

# Fill missing cast with 'Unknown'
df_clean['cast'] = df_clean['cast'].fillna('Unknown')

# Fill missing country with 'Unknown'
df_clean['country'] = df_clean['country'].fillna('Unknown')

# Fill missing rating with 'Not Rated'
df_clean['rating'] = df_clean['rating'].fillna('Not Rated')

# Convert date_added to datetime
df_clean['date_added'] = pd.to_datetime(df_clean['date_added'], errors='coerce')

# Extract year and month from date_added
df_clean['year_added'] = df_clean['date_added'].dt.year
df_clean['month_added'] = df_clean['date_added'].dt.month
df_clean['month_name'] = df_clean['date_added'].dt.month_name()

# Create duration numeric column
df_clean['duration_numeric'] = df_clean['duration'].str.extract('(\d+)').astype('float')

# Create content age categories
def categorize_content_age(year):
    if pd.isna(year):
        return 'Unknown'
    elif year >= 2020:
        return 'Very Recent (2020+)'
    elif year >= 2015:
        return 'Recent (2015-2019)'
    elif year >= 2010:
        return 'Modern (2010-2014)'
    elif year >= 2000:
        return 'Early 2000s'
    else:
        return 'Classic (Pre-2000)'

df_clean['content_age_category'] = df_clean['release_year'].apply(categorize_content_age)

print("Data wrangling completed!")
print(f"Clean dataset shape: {df_clean.shape}")

````

### What all manipulations have you done and insights you found?

1.  **Creating a copy for cleaning** : df_clean = df.copy()


*  **Manipulation**: This creates a separate copy of the original DataFrame (df) to perform cleaning operations on. This is a good practice as it preserves the original data in case you need to refer back to it.
*  **Insight**: This is a preparatory step for cleaning and doesn't provide direct insights into the data itself.


2.   **Handling missing values**: fillna() with 'Unknown' or 'Not Rated'


*   **Manipulation**: Missing values in director, cast, country, and rating are replaced with placeholder strings ('Unknown' or 'Not Rated'). This is a simple imputation strategy suitable when the missingness might represent unknown information rather than a meaningful zero or average.
*   **Insight**: The proportion of missing values in these columns (seen in the "Missing Values Analysis") tells you how much information is unavailable for these attributes. Filling them allows you to include these rows in analysis without dropping them, although the 'Unknown'/ 'Not Rated' category should be interpreted cautiously.


3.   **Converting date_added to datetime**: pd.to_datetime(df_clean['date_added'], errors='coerce')


*   **Manipulation**: The date_added column, initially stored as strings, is converted into datetime objects. errors='coerce' will turn any values that cannot be parsed into datetime objects into NaT (Not a Time).
*   **Insight**: This conversion is essential for performing time-based analysis, such as tracking content addition trends over time.


4.   **Extracting year and month from date_added**: .dt.year, .dt.month, .dt.month_name()

*   **Manipulation**: New columns year_added, month_added, and month_name are created by extracting the year, numerical month, and month name from the date_added column.
*   **Insight**: These columns allow you to analyze seasonal patterns in content addition or track trends on a yearly or monthly basis.
5.  **Creating duration numeric column**: .str.extract('(\d+)').astype('float')
*     **Manipulation**: This extracts the numerical part (the number of minutes for movies or seasons for TV shows) from the duration string column and converts it to a float. This is done using regular expressions to find digits within the string.
*      **Insight**: This creates a quantitative measure of duration, which is necessary for numerical analysis and visualizations related to content length. It also highlights the need to differentiate between movie durations (in minutes) and TV show durations (in seasons) when analyzing this column.

6. **Creating content age categories**: categorize_content_age function and .apply()

*  **Manipulation**: A new categorical column content_age_category is created based on the release_year. Content is grouped into categories like 'Very Recent', 'Recent', 'Modern', etc.
*  **Insight**: This simplifies the analysis of content age by grouping titles into meaningful time periods. It allows you to examine how the volume or characteristics of content vary across different age groups.
